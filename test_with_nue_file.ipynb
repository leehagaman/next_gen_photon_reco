```python
import uproot
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = uproot.open("../epem_gamma_consistency/spacepoint_rootfiles/run_3/nue_14338_312_15601.root")

f.keys()
```

```python
rse_df = f["Trun"].arrays(["runNo", "subRunNo", "eventNo"], library="pd")
track_shower_spacepoints_df = f["T_rec"].arrays(["x", "y", "z", "q"], library="pd")
spacepoints_blob_df = f["T_rec_charge_blob"].arrays(["x", "y", "z", "q"], library="pd")
spacepoints_cosmic_df = f["T_cluster"].arrays(["x", "y", "z", "q"], library="pd")

print("\nrse_df")
display(rse_df)
print("\nrack_shower_spacepoints_df")
display(track_shower_spacepoints_df)
print("\nspacepoints_blob_df")
display(spacepoints_blob_df)
print("\nspacepoints_cosmic_df")
display(spacepoints_cosmic_df)

```

```python
plt.figure()
plt.hist(track_shower_spacepoints_df["q"], bins=11, histtype="step", density=True, label="spacepoints")
plt.hist(spacepoints_blob_df["q"], bins=100, histtype="step", density=True, label="spacepoints_blob")
plt.hist(spacepoints_cosmic_df["q"], bins=100, histtype="step", density=True, label="spacepoints_cosmic")
plt.legend()
plt.xlabel("q")
plt.ylabel("density")
plt.show()
```

```python
s = 1

plt.figure()
plt.scatter(track_shower_spacepoints_df["x"], track_shower_spacepoints_df["y"], c=track_shower_spacepoints_df["q"], s=1, label="spacepoints", cmap="jet")
plt.colorbar()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("track_shower_spacepoints")
plt.show()

plt.figure()
plt.scatter(spacepoints_blob_df["x"], spacepoints_blob_df["y"], c=spacepoints_blob_df["q"], s=1, label="spacepoints_blob", cmap="jet")
plt.colorbar()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("spacepoints_blob")
plt.show()

plt.figure()
plt.scatter(spacepoints_cosmic_df["x"], spacepoints_cosmic_df["y"], c=spacepoints_cosmic_df["q"], s=1, label="spacepoints_cosmic", cmap="jet")
plt.colorbar()
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("spacepoints_cosmic")
plt.show()


```
